```
import urllib.request
import numpy as np
from matplotlib import pyplot as plt

dataset = urllib.request.urlopen("https://raw.githubusercontent.com/Tabrizian/ml-homework/master/HW1/dataset/data1.txt").read().decode('ascii')
```

```
dataset = dataset.split('\n')
for index in range(len(dataset)):
  dataset[index] = dataset[index].split(',')
```

```
train_size = int(len(dataset) * 0.6)
eval_size = int(len(dataset) * 0.2)
test_size = int(len(dataset) * 0.2)

training_set = dataset[0:train_size]
evaluation_set = dataset[train_size: train_size + eval_size]
test_set = dataset[train_size + eval_size: len(dataset) - 1]

X_train = [float(x[0]) for x in training_set]
Y_train = np.transpose([float(y[1]) for y in training_set]).reshape((-1,1))

X_eval = [float(x[0]) for x in evaluation_set]
Y_eval = np.transpose([float(y[1]) for y in evaluation_set]).reshape((-1,1))

X_test = [float(x[0]) for x in test_set]
Y_test = np.transpose([float(y[1]) for y in test_set]).reshape((-1,1))
```

```
def get_features(X):
  X_features = [np.power(X, i) for i in range(0, 6)]
  return np.transpose(X_features)
```

```
def calculate_cost(X, y, theta):
  m = len(X)
  X = get_features(X)
  predicted = np.matmul(X, theta)
  return np.divide(np.sum(np.power(np.subtract(y, predicted), 2)), 2 * m)
  
  
```

```
def calculate_gradient(X, theta, alpha, y):
  m = len(X)
  X = get_features(X)
  predicted = np.matmul(X, theta)

  difference = predicted - y
  return np.multiply(np.divide(np.matmul(np.transpose(X), difference), m), alpha)
  
```

```
def gradient_descent(X, theta, y, alpha, num_iter):
  cost_history = []
  for i in range(num_iter):
    theta = np.subtract(theta, calculate_gradient(X, theta, alpha, y))
    cost_history.append(calculate_cost(X, y, theta))
  return theta, cost_history
```

```

def PolyCoefficients(x, coeffs):

  """ Returns a polynomial for ``x`` values for the ``coeffs`` provided.
  The coefficients must be in ascending order (``x**0`` to ``x**o``).
  """
  o = len(coeffs)
  print(f'# This is a polynomial of order {o - 1}.')
  y = 0
  for i in range(o):
    y += coeffs[i]*x**i
  return y

```

```
def plot(x, y, theta):
  x = np.asarray(x)
  plt.scatter(x, y, s=80, marker=">")
  plt.plot(x, PolyCoefficients(x, theta), color='red')
  plt.show()
  
```

```
def plotCostHistory(cost_history):
  plt.plot(cost_history, 'ro')
  
  plt.show()
  
```

```
theta_1, cost_history_1 = gradient_descent(X_train, [[0], [0], [0], [0], [0], [0]], Y_train, 0.8, 10)
theta_2, cost_history_2 = gradient_descent(X_train, [[0], [0], [0], [0], [0], [0]], Y_train, 0.8, 100)
theta_3, cost_history_3 = gradient_descent(X_train, [[0], [0], [0], [0], [0], [0]], Y_train, 0.8, 1000)

plot(X_train, Y_train, theta_1)
plot(X_train, Y_train, theta_2)
plot(X_train, Y_train, theta_3)
plotCostHistory(cost_history_1)
plotCostHistory(cost_history_2)
plotCostHistory(cost_history_3)

```
